## All teams

```python
teams = matches["team1"].unique()
teams
```

```python
totalrunsofeachteamarr = []
for i in teams :
    temp = deliveries.loc[(deliveries["batting_team"] == i),["batting_team","total_runs"]]
    tempruns = temp["total_runs"].count()
    totalrunsofeachteamarr.append(tempruns)
totalrunsofeachteamarr
```

## Runs scored by each team

```python
totalrunsofeachteam = pd.DataFrame({"Team" : teams,"Total runs of each team" : totalrunsofeachteamarr}, columns = ["Team", "Total runs of each team"]).set_index("Team")
totalrunsofeachteam.sort_values(by="Total runs of each team", ascending=False,inplace = True)
totalrunsofeachteam
```

## Result : Highest number of runs are scored by Mumbai Indians

```python
plt.figure(figsize=(14,14))
sns.barplot(totalrunsofeachteam["Total runs of each team"],totalrunsofeachteam.index, palette ="deep")
```

```python
batsman = deliveries["batsman"].unique()
batsman
```

```python
totalrunseachbatsmanarr = []
for i in batsman :
    temp1 = deliveries.loc[deliveries["batsman"]==i,["batsman","batsman_runs"]]
    temp2 = temp1["batsman_runs"].sum()
    totalrunseachbatsmanarr.append(temp2)
totalrunseachbatsmanarr
```

```python
totalrunseachbatsman = pd.DataFrame({"Batsman" : batsman,"Total runs by each Batsman" : totalrunseachbatsmanarr}, columns = ["Batsman", "Total runs by each Batsman"]).set_index("Batsman")
totalrunseachbatsman.sort_values(by="Total runs by each Batsman", ascending = False,inplace=True)
top25batsmanruns = totalrunseachbatsman["Total runs by each Batsman"].nlargest(25)
top25batsmanruns
```

```python
plt.figure(figsize=(12,12))
sns.barplot(top25batsmanruns,top25batsmanruns.index, palette ="hls")
plt.tight_layout()
```

```python
plt.figure()
sns.barplot(sixescount,sixescount.index, palette ="Blues_r", orient="h")
plt.tight_layout()
```

```python
eachbowlerwickets = []
for i in bowlers :    
    temp10 = deliveries.loc[(pd.isna(deliveries["player_dismissed"])==False)&(deliveries["bowler"]==i)&(deliveries["dismissal_kind"]!="run out")&(deliveries["dismissal_kind"]!="retired hurt")&(deliveries["dismissal_kind"]!="obstructing the field"),["bowler","player_dismissed"]]
    temp11 = temp10.shape[0]
    eachbowlerwickets.append(temp11)
df_eachbowlerwickets = pd.DataFrame({"Bowler" : bowlers,"Total wickets" : eachbowlerwickets}, columns = ["Bowler", "Total wickets"]).set_index("Bowler")
df_eachbowlerwickets.sort_values(by="Total wickets", ascending = False, inplace = True)
df_eachbowlerwickets
```

## Top Bowler : SL Malinga

```python
maxm = 0
match_ids = deliveries["match_id"].unique()
for i in match_ids :
    temp22 = deliveries.loc[deliveries["match_id"]==i,:].groupby(["match_id","batting_team","bowling_team","bowler"])["over"].value_counts().max()
    if temp22 > maxm :
        maxm = temp22
print(maxm)
```
